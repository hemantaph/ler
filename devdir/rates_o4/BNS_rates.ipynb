```python
import numpy as np
import matplotlib.pyplot as plt
# call the LeR class
from ler.rates import LeR
from ler.utils import get_param_from_json
```

```python
# LeR initialization with default arguments
ler = LeR(
    npool=6,
    event_type='BNS',
    z_max=10,
    mtot_max=50.0,
    mtot_min=1.0,
    verbose=False,
    )
```

```python
ler.selecting_n_unlensed_detectable_events(
    size=100,
    batch_size=200000,
    snr_threshold=8.0,
    resume=True,
    output_jsonfile='n_unlensed_param_detectable_BNS_spinless.json',
    meta_data_file='meta_unlensed_BNS_spinless.json',
    detectability_condition='step_function',
    trim_to_size=False,
);
```

```python
meta_data= get_param_from_json("./ler_data/meta_unlensed_BNS_spinless.json")

# plot the rate vs sampling size
plt.figure(figsize=(4,4))
plt.plot(meta_data['events_total'], meta_data['total_rate'], 'o-')
plt.xlabel('Sampling size')
plt.ylabel('Rate (per year)')
plt.title('Rate vs Sampling size')
plt.grid(alpha=0.4)
plt.show()
```

```python
unlensed_param = get_param_from_json("./ler_data/n_unlensed_param_detectable_BNS_spinless.json")
# select only events after sampling size of 
idx = np.where(meta_data['events_total'] > 3e6)[0]
# take average 
unlensed_rate = np.mean(meta_data['total_rate'][idx])
print('Unlensed GW Rate (per year), SNR>8 =', unlensed_rate)
```

```python
ler.selecting_n_lensed_detectable_events(
    size=100,
    batch_size=200000,
    snr_threshold=[8.0, 8.0],
    num_img=[1, 1],
    resume=True,
    detectability_condition='step_function',
    output_jsonfile='n_lensed_params_detectable_BNS_spinless.json',
    meta_data_file='meta_lensed_BNS_spinless.json',
    trim_to_size=False,
    nan_to_num=False,
)
```
