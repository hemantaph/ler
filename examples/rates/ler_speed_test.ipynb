## NO-NJIT

```python
# disable numba jit 
import os
os.environ["NUMBA_DISABLE_JIT"] = "1"

# set number of threads to 1
import numba as nb
nb.set_num_threads(1)   # caps Numba parallel regions to 1 thread

# set multiprocessing npool to 1
npool = 1
```

```python
from ler.rates import LeR
import numpy as np
import time

# Uncomment the below code if you need to change the default arguments.
ler_nonjit = LeR(
    npool=1, # number of processors to use
    interpolator_directory='../interpolator_json', 

    lens_functions = dict(
        cross_section_based_sampler = 'rejection_sampler_partial',
        cross_section = 'cross_section_epl_shear_njit',
    ),
    image_properties_function = 'image_properties_epl_shear_lenstronomy',
)
```

```python
start_time_nonjit = time.time()
unlensed_param = ler_nonjit.unlensed_cbc_statistics(size=10000, batch_size=10000, resume=False, output_jsonfile=False)
# unlensed_rate, unlensed_param_detectable = ler_nonjit.unlensed_rate(unlensed_param, output_jsonfile=False)
end_time_nonjit_unlensed = time.time()
print(f"\nTime taken for unlensed_cbc_statistics: {end_time_nonjit_unlensed - start_time_nonjit} seconds\n")

lensed_param = ler_nonjit.lensed_cbc_statistics(size=50, batch_size=50, resume=False, output_jsonfile=False)
# lensed_rate, lensed_param_detectable = ler_nonjit.lensed_rate(lensed_param, output_jsonfile=False)
end_time_nonjit_lensed = time.time()
print(f"\nTime taken for lensed_cbc_statistics: {end_time_nonjit_lensed - end_time_nonjit_unlensed} seconds\n")

# total time taken
print(f"\nTotal time taken: {end_time_nonjit_lensed - start_time_nonjit} seconds\n")
```

## NJIT

```python
from ler.rates import LeR
import numpy as np
import time

# Uncomment the below code if you need to change the default arguments.
ler_njit = LeR(
    npool=6, # number of processors to use
    interpolator_directory='../interpolator_json', 
)
```

```python
# warm-up
unlensed_param = ler_njit.unlensed_cbc_statistics(size=2, batch_size=2, resume=False, output_jsonfile=False)
# unlensed_rate, unlensed_param_detectable = ler_njit.unlensed_rate(unlensed_param, output_jsonfile=False)

lensed_param = ler_njit.lensed_cbc_statistics(size=2, batch_size=2, resume=False, output_jsonfile=False)
# lensed_rate, lensed_param_detectable = ler_njit.lensed_rate(lensed_param, output_jsonfile=False)
```

```python
# warm-up
start_time_njit = time.time()
unlensed_param = ler_njit.unlensed_cbc_statistics(size=10000, batch_size=10000, resume=False, output_jsonfile=False)
# unlensed_rate, unlensed_param_detectable = ler_njit.unlensed_rate(unlensed_param, output_jsonfile=False)
end_time_njit_unlensed = time.time()
print(f"\nTime taken for unlensed_cbc_statistics: {end_time_njit_unlensed - start_time_njit} seconds\n")

lensed_param = ler_njit.lensed_cbc_statistics(size=50, batch_size=50, resume=False, output_jsonfile=False)
# lensed_rate, lensed_param_detectable = ler_njit.lensed_rate(lensed_param, output_jsonfile=False)
end_time_njit_lensed = time.time()
print(f"\nTime taken for lensed_cbc_statistics: {end_time_njit_lensed - end_time_njit_unlensed} seconds\n")

# total time taken
print(f"\nTotal time taken: {end_time_njit_lensed - start_time_njit} seconds\n")
```
