```python
import corner
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
from ler.utils import get_param_from_json

# Load data
ann = get_param_from_json('./ler_data/unlensed_params_detectable_ann.json')
innerp = get_param_from_json('./ler_data/unlensed_params_detectable_BBH_innerp.json')
interp = get_param_from_json('./ler_data/unlensed_params_detectable_BBH_intrp.json')
```

```python
ann.keys()
```

```python
# Parameters to compare
param_names = ['zs', 'mass_1_source', 'mass_2_source', 'a_1', 'a_2', 'tilt_1', 'tilt_2', 'phi_12', 'phi_jl']
labels = ['$z_s$', '$m_1$ [$M_\odot$]', '$m_2$ [$M_\odot$]', 'a_1', 'a_2', 'tilt_1', 'tilt_2', 'phi_12', 'phi_jl']

# Create DataFrames
size = 5000
size = np.min([len(ann[param_names[0]]), len(innerp[param_names[0]]), len(interp[param_names[0]])])
indices = np.random.choice(len(ann[param_names[0]]), size=size, replace=False)


df_ann = pd.DataFrame({label: ann[name][indices] for name, label in zip(param_names, labels)})
df_ann['Event Type'] = 'ann'
df_innerp = pd.DataFrame({label: innerp[name][indices] for name, label in zip(param_names, labels)})
df_innerp['Event Type'] = 'inner_product'
df_intrp = pd.DataFrame({label: interp[name][indices] for name, label in zip(param_names, labels)})
df_intrp['Event Type'] = 'interpolation_aligned_spins'

# Combine DataFrames
df_combined = pd.concat([df_ann, df_innerp, df_intrp], ignore_index=True)

# Calculate 1st and 99th percentiles for each parameter to set limits
percentile_limits = {}
for label in labels:
    lower = np.percentile(df_combined[label], 1)
    upper = np.percentile(df_combined[label], 99)
    # Add some padding (5% of range)
    range_pad = (upper - lower) * 0.05
    percentile_limits[label] = (lower - range_pad, upper + range_pad)

# Prepare corner plot dataframes (use same sample size for each)
size = np.min([len(ann[param_names[0]]), len(innerp[param_names[0]]), len(interp[param_names[0]])])
indices = np.random.choice(len(ann[param_names[0]]), size=size, replace=False)
samples_all = np.stack([ann[p][indices] for p in param_names], axis=1)
samples_innerp = np.stack([innerp[p][indices] for p in param_names], axis=1)
samples_intrp = np.stack([interp[p][indices] for p in param_names], axis=1)
# Use the previously defined `labels` which includes tilt_2, phi_12, phi_jl
labels_corner = labels

# Create corner plot for all events
fig = corner.corner(
    samples_all,
    labels=labels_corner,
    color='C0',
    alpha=0.5,
    plot_density=False,
    plot_datapoints=False,
    smooth=0.8,
    hist_kwargs={'density': True}
)

# Overlay inner_product and interpolation_aligned_spins samples
corner.corner(samples_innerp, labels=labels_corner, color='C1', alpha=0.5, fig=fig, plot_density=False, plot_datapoints=False, smooth=0.8, hist_kwargs={'density': True})
corner.corner(samples_intrp, labels=labels_corner, color='C2', alpha=0.5, fig=fig, plot_density=False, plot_datapoints=False, smooth=0.8, hist_kwargs={'density': True})

# Add legend
blue_line = mlines.Line2D([], [], color='C0', label='ann')
orange_line = mlines.Line2D([], [], color='C1', label='inner_product')
green_line = mlines.Line2D([], [], color='C2', label='interpolation_aligned_spins')
fig.legend(handles=[blue_line, orange_line, green_line], loc='upper right', bbox_to_anchor=(0.95, 0.95), fontsize=12)
fig.suptitle('Unlensed Events: All vs Detectable (corner)', fontsize=16, y=1.02)
plt.show()
```
