```python
from ler.utils import generate_mixed_grid
import numpy as np
import matplotlib.pyplot as plt 

# default
power_law_grid = generate_mixed_grid(
    x_min=0.001,
    x_max=10.0,
    resolution=20,
    power_law_part='lower',
    geomspace_part=False,
    spacing_trend='increasing',
    power=2.3,
    value_transition_fraction=0.6,
    num_transition_fraction=0.8,
    auto_match_slope=True,
)
geomspace_grid = generate_mixed_grid(
    x_min=0.001,
    x_max=10.0,
    resolution=20,
    geomspace_part='lower',
    spacing_trend='increasing',
    value_transition_fraction=0.6,
    num_transition_fraction=0.8,
    auto_match_slope=True,
)

# plot to see spacing
plt.plot(power_law_grid, '.', color='C0', label='power-law')    
plt.plot(geomspace_grid, '.', color='C1', label='geomspace')
plt.legend()
plt.show()
```
